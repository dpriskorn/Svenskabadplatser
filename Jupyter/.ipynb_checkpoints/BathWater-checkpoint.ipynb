```python
import pandas as pd
import csv
import os


cwd = os.getcwd()
print (cwd)
#data = pd.read_csv("baths.csv")
df = pd.read_csv('baths.csv', error_bad_lines=False,delimiter="|",warn_bad_lines=False)
df.info()
```

```python
import seaborn as sns
sns.set_theme()
sns.set(rc={'figure.figsize':(20,10)})
dfg = df.groupby('countryCode').count()
dfg.reset_index(inplace=False)
dfg
```

```python
sns.barplot(y=dfg.bathingWaterIdentifier,x=dfg.index, data=dfg);
```

```python
dfgBW = dfg.sort_values(by=['bathingWaterIdentifier'])
sns.barplot(y=dfgBW.bathingWaterIdentifier,x=dfgBW.index, data=dfgBW)
```
